Sets

```python
#What is unique about a set?
# a set can contain only unique values and not any repeated values, it store the values in curly braces but without any colons that separates it form dictionary as a set does not work on key:value pairs
```

```python
#Use a set to find the unique values of the list below:
list5 = [1,2,2,33,4,4,11,22,3,3,2]
set(list5)
```

```python
data = [1, "1", 2, "2", 2, "02", 3, "3", "03", 1]
unique_values = set(int(x) for x in data)

unique_values
set(unique_values)
```
